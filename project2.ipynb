```python
import pandas as pd
import numpy as np
import pyodbc
```

```python
luxury=pd.read_csv("Luxury_Housing_Bangalore.csv")
```

```python
luxury.head(5)
```

```python
luxury.columns
```

```python
luxury['Property_ID'].unique()
```

```python
luxury['Micro_Market'].unique()
```

```python
luxury["Micro_Market"] = luxury["Micro_Market"].str.lower().str.title()
```

```python
luxury['Micro_Market'].unique()
```

```python
luxury['Project_Name'].unique()
```

```python
luxury['Developer_Name'].unique()
```

```python
luxury['Developer_Name'] = luxury['Developer_Name'].str.replace("&", "", regex=False)
```

```python
luxury['Unit_Size_Sqft'].unique()
```

```python
luxury['Unit_Size_Sqft'] = luxury['Unit_Size_Sqft'].astype('float')
```

```python
luxury['Unit_Size_Sqft'] = luxury['Unit_Size_Sqft'].fillna(luxury['Unit_Size_Sqft'].median()).astype('float')
```

```python
luxury['Unit_Size_Sqft'].unique()
```

```python
luxury['Configuration'].unique()
```

```python
luxury['Configuration'] = luxury['Configuration'].str.upper()
```

```python
luxury['Configuration'] = luxury['Configuration'].str.replace(" ", "")
```

```python
luxury['Configuration'] = luxury['Configuration'].str.replace(r"(\d)BHK", r"\1 BHK", regex=True)
```

```python
luxury['Configuration'] = luxury['Configuration'].str.replace("+", "", regex=False)
```

```python
luxury['Configuration'].unique()
```

```python
luxury['Ticket_Price_Cr'].unique()
```

```python
luxury['Ticket_Price_Cr'] = (luxury['Ticket_Price_Cr'].str.replace('₹', '', regex=False).str.replace('Cr', '', regex=False).str.strip())
```

```python
luxury['Ticket_Price_Cr'] = luxury['Ticket_Price_Cr'].astype(float)
```

```python
luxury['Ticket_Price_Cr'] = luxury['Ticket_Price_Cr'].fillna(luxury['Ticket_Price_Cr'].median()).astype('float')
```

```python
luxury['Ticket_Price_Cr'] = luxury['Ticket_Price_Cr'].round(2)
```

```python
luxury['Transaction_Type'].unique()
```

```python
luxury['Buyer_Type'].unique()
```

```python
luxury['Purchase_Quarter'].unique()
```

```python
luxury['Purchase_Quarter'] = pd.to_datetime(luxury['Purchase_Quarter'])
```

```python
luxury['Purchase_Quarter'].dtype
```

```python
luxury['Connectivity_Score'].unique()
```

```python
luxury['Connectivity_Score'] = luxury['Connectivity_Score'].fillna(luxury['Connectivity_Score'].median()).astype('float')
```

```python
luxury['Connectivity_Score'] = luxury['Connectivity_Score'].round(2)
```

```python
luxury['Amenity_Score'].unique()
```

```python
luxury['Amenity_Score'] = luxury['Amenity_Score'].round(2)
```

```python
luxury['Amenity_Score'] = luxury['Amenity_Score'].fillna(luxury['Amenity_Score'].median()).astype('float')
```

```python
luxury['Possession_Status'].unique()
```

```python
luxury['Sales_Channel'].unique()
```

```python
luxury['NRI_Buyer'].unique()
```

```python
luxury['Locality_Infra_Score'].unique()
```

```python
luxury['Locality_Infra_Score'] = luxury['Locality_Infra_Score'].round(2)
```

```python
luxury['Avg_Traffic_Time_Min'].unique()
```

```python
luxury['Buyer_Comments'].unique()
```

```python
luxury['Buyer_Comments'] = luxury['Buyer_Comments'].astype(str)
```

```python
luxury['Buyer_Comments'] = luxury['Buyer_Comments'].fillna('Unknown')
```
